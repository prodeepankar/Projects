Ques 2) Which language has the average highest ROI ?

```python
def Return_on_investment(Profit,Budget):
  if Budget > 0:
    ROI = (Profit/Budget)*100
  else:
    ROI = pd.NaN 
  return ROI
df['Return on investment'] = df.apply(lambda x: Return_on_investment(x['Updated_Profit'],x['budget']),axis=1)
print(df[['budget','Updated_Profit','Return on investment']])
```

```python
def list_producers(index):
  no_null_crew = df[~df['crew'].isnull()].iloc[index].loc['crew']
  list_of_producers = []
  for crew in no_null_crew:
    if crew['job'] == 'Producer':
      list_of_producers.append(crew['name'])
  return list_of_producers
```

```python
def list_directors(index):
  no_null_crew = no_null_crew = df[~df['crew'].isnull()].iloc[index].loc['crew']
  director_list = []
  for crew in no_null_crew:
    if crew['job'] == 'Director':
        director_list.append(crew['name'])
  return director_list
```

Ques 5) What is the actor has acted in most movies ?

```python
df['cast'].loc[~df['cast'].isnull()] = df['cast'].loc[~df['cast'].isnull()].apply(convert_str_to_list)
no_null_cast = df.loc[~df['cast'].isnull()]
```

```python
actors_list = []
for index,value in no_null_cast.iterrows():
  cast = no_null_cast.loc[index,'cast']
  for k in cast:
    if 'character' in k:
        actors_list.append(k['name'])
print(actors_list)
```

```python
Actor_Table = pd.DataFrame(actors_list,columns=['Name of Actor'])
```

```python
Actor_Table.value_counts().reset_index().head(4)
```
